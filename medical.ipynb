```python
# prompt: generate graph from the above data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'insurance.csv' is in your current directory or you provide the correct path
# If the file is not present, this will raise a FileNotFoundError
data = pd.read_csv('/content/insurance.csv')


# Create a pairplot to visualize relationships between numerical variables
sns.pairplot(data, hue='smoker')
plt.show()


# Create a boxplot to show the distribution of charges across different regions
plt.figure(figsize=(10, 6))
sns.boxplot(x='region', y='charges', data=data, palette='viridis')
plt.title('Distribution of Charges across Regions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()


# Create a violin plot to show the distribution of charges for smokers and non-smokers
plt.figure(figsize=(8, 6))
sns.violinplot(x='smoker', y='charges', data=data, palette='pastel')
plt.title('Distribution of Charges for Smokers vs. Non-Smokers')
plt.show()
```

```python
# prompt: generate 3D plot for above data

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming 'insurance.csv' is in your current directory or you provide the correct path
data = pd.read_csv('/content/insurance.csv')

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Assuming 'age', 'bmi', and 'charges' are the columns you want to plot
x = data['age']
y = data['bmi']
z = data['charges']

ax.scatter(x, y, z, c=z, cmap='viridis')  # Use charges for color mapping

ax.set_xlabel('Age')
ax.set_ylabel('BMI')
ax.set_zlabel('Charges')
ax.set_title('3D Scatter Plot of Age, BMI, and Charges')

plt.show()
```
