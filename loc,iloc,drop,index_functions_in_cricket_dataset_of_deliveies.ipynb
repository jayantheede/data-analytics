```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/deliveries.csv', index_col=9)
df.head()
```

```python
subset = df.loc[[7, 10], ['batsman', 'bowler']]
print(subset)
```

```python
batsman_indexer = list(range(5))
coloumn_indexer = [df.columns.get_loc('batsman'), df.columns.get_loc('bowler')]

```

```python
subset = df.iloc[batsman_indexer, coloumn_indexer]
```

```python
print(subset)
```

```python
import pandas as pd

df = pd.read_csv('/content/deliveries.csv', index_col=9)
df = df.reset_index()
if 'Unnamed: 9' in df.columns:
    df.drop(columns=['Unnamed: 9'], inplace=True)
elif '9' in df.columns:
    df.drop(columns=['9'], inplace=True)
else:
    print("Column '9' not found in the DataFrame.")
df = df.drop(index=[7, 10], errors='ignore')
print(df.head())
```

```python
print(df.tail())
```

```python

if df.columns.size > 9:
  df.set_index(df.columns[9],inplace=True)
df = df.drop(index=[7, 10], errors='ignore')
print(df.head())
print(df.tail())
```

```python

if df.columns.size > 9:
  df.set_index(df.columns[9],inplace=False)
df = df.drop(index=[7, 10], errors='ignore')
print(df.head())
print(df.tail())
```
