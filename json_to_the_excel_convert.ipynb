```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd

# Example dataset (replace with your actual data)

data = {

    'timestamp': pd.date_range(start='2020-01-01', periods=600, freq='D'),  # 600 days starting from Jan 1, 2020

    'bike_shares': [100 + (x*5) for x in range(600)],  # Example bike shares data (increasing over time)

    'holiday': [1 if x % 15 == 0 else 0 for x in range(600)],  # Simulate holidays every 15th day

    'season': ['Winter', 'Spring', 'Summer', 'Fall'] * 150  # Simulated seasons for simplicity

}

# Create DataFrame

df = pd.DataFrame(data)

# Extract year from the timestamp

df['year'] = df['timestamp'].dt.year

# 1. Plotting Bike Shares by Year

plt.figure(figsize=(10, 6))

# Group by year and calculate the total bike shares for each year

df_yearly = df.groupby('year')['bike_shares'].sum().reset_index()

# Plotting the data

sns.barplot(x='year', y='bike_shares', data=df_yearly, palette='viridis')

plt.title('Total Bike Shares by Year')

plt.xlabel('Year')

plt.ylabel('Total Bike Shares')

# Show plot

plt.tight_layout()

plt.show()

# 2. Plotting Bike Shares on Holidays by Season

plt.figure(figsize=(10, 6))

# Filter data for holidays

df_holidays = df[df['holiday'] == 1]

# Group by season and calculate total bike shares on holidays

df_season_holiday = df_holidays.groupby('season')['bike_shares'].sum().reset_index()

# Plotting the data

sns.barplot(x='season', y='bike_shares', data=df_season_holiday, palette='Set2')

plt.title('Bike Shares on Holidays by Season')

plt.xlabel('Season')

plt.ylabel('Total Bike Shares on Holidays')

# Show plot

plt.tight_layout()

plt.show()

```
