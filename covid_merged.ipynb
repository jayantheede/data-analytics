```python
import pandas as pd  # Importing pandas library

# Load your DataFrames (update file paths)
df1 = pd.read_csv('/content/covid_merged.csv')  # Replace with actual file path
df2 = pd.read_csv('/content/merged_date_based.csv')  # Replace with actual file path

# Convert 'Latitude' column in df1 to float64
df1['Latitude'] = pd.to_numeric(df1['Latitude'], errors='coerce')

# Convert 'Latitude' column in df2 to float64 (to ensure matching data types)
df2['Latitude'] = pd.to_numeric(df2['Latitude'], errors='coerce')

# Drop NaN values in 'Latitude' if necessary
df1 = df1.dropna(subset=['Latitude'])
df2 = df2.dropna(subset=['Latitude'])

# Merge on 'Latitude' column
merged_df = pd.merge(df1, df2, on='Latitude', how='inner')  # Use 'inner' join to keep matching rows

# Save merged DataFrame
merged_df.to_csv('merged_file.csv', index=False)

print("Merge completed. File saved as 'merged_file.csv'.")
```
