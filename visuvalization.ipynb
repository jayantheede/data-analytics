```python
# prompt: Altair

import altair as alt
import pandas as pd

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Display the first few rows
print(df.head())

# Display the last few rows
print(df.tail())

# Reset the index for proper visualization if needed
df = df.reset_index()
# Altair chart
chart = alt.Chart(df).mark_line().encode(
    x='index',  # Assuming you want to plot against the index
    y=alt.Y(df.columns[1], type='quantitative'), # Assuming you want to plot the second column
    tooltip=['index', df.columns[1]] # Add tooltips if needed, modify columns as needed.
).properties(
    title='Data Visualization with Altair'
)

chart
```

```python
# prompt: ggplot

!pip install plotnine
import pandas as pd
from plotnine import ggplot, aes, geom_line, labs

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Reset the index for proper visualization if needed
df = df.reset_index()

# ggplot chart
plot = (
    ggplot(df, aes(x='index', y=df.columns[1]))  # Assuming you want to plot against the index and the second column
    + geom_line()
    + labs(title='Data Visualization with ggplot', x='Index', y=df.columns[1])
)

# Display the plot
plot
```

```python
# prompt: Bokeh

import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Reset the index for proper visualization
df = df.reset_index()

# Output to notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="Data Visualization with Bokeh", x_axis_label='Index', y_axis_label=df.columns[1])

# Add a line renderer
p.line(df['index'], df[df.columns[1]], line_width=2)

# Add hover tool
hover = HoverTool()
hover.tooltips = [
    ("Index", "$x"),
    (df.columns[1], "$y")
]
p.add_tools(hover)

# Show the results
show(p)
```

```python
# prompt: Plotly

import pandas as pd
import plotly.express as px

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Display the first few rows
print(df.head())

# Display the last few rows
print(df.tail())

# Reset the index
df = df.reset_index()

# Plotly Express line chart
fig = px.line(df, x='index', y=df.columns[1], title='Data Visualization with Plotly Express',
                labels={'index': 'Index', df.columns[1]: df.columns[1]})
fig.show()
```

```python
# prompt: Seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Display the first few rows
print(df.head())

# Display the last few rows
print(df.tail())

# Reset the index
df = df.reset_index()

# Seaborn line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='index', y=df.columns[1], data=df)
plt.title('Data Visualization with Seaborn')
plt.xlabel('Index')
plt.ylabel(df.columns[1])
plt.show()
```

```python
# prompt: Matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Reset the index
df = df.reset_index()

# Matplotlib line plot
plt.figure(figsize=(10, 6))
plt.plot(df['index'], df[df.columns[1]])
plt.title('Data Visualization with Matplotlib')
plt.xlabel('Index')
plt.ylabel(df.columns[1])
plt.grid(True)
plt.show()
```

```python
# prompt: Holoviews

!pip install holoviews
import pandas as pd
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Reset the index
df = df.reset_index()

# Create a HoloViews line plot
line_plot = hv.Curve(df, kdims='index', vdims=df.columns[1])

# Customize the plot (optional)
line_plot = line_plot.opts(
    title="Data Visualization with HoloViews",
    xlabel="Index",
    ylabel=df.columns[1],
    width=800,
    height=400,
    line_width=2,
    tools=['hover'],  # Enable hover tool

)

# Display the plot
line_plot
```

```python
# prompt: scatter plot

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Reset the index
df = df.reset_index()

# Scatter plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(df['index'], df[df.columns[1]])
plt.title('Scatter Plot of Data')
plt.xlabel('Index')
plt.ylabel(df.columns[1])
plt.grid(True)
plt.show()
```

```python
# prompt: histogram

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataframe
df = pd.read_csv('/content/data (1).csv')

# Reset the index
df = df.reset_index()

# Histogram using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df[df.columns[1]], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel(df.columns[1])
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
