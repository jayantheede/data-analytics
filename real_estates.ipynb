# **2. Data Cleaning and Preprocessing**
1)Handle missing values:  


*Impute numerical columns with median.


*Impute categorical columns with mode.

```python
#  Handle missing values:
# Impute numerical columns with median.

# Assuming 'data' DataFrame is already loaded as in the previous code.

# Impute numerical columns with the median
for col in data.select_dtypes(include=np.number):
    if data[col].isnull().any():
        median_val = data[col].median()
        data[col].fillna(median_val, inplace=True)

print("\nMissing Values after imputation:")
print(data.isnull().sum())
```

This code identifies missing values in numerical columns and replaces them with the median of each column.




```python
# prompt: Handle missing values:
# Impute categorical columns with mode.

# Impute categorical columns with the mode
for col in data.select_dtypes(include='object'):
    if data[col].isnull().any():
        mode_val = data[col].mode()[0]  # Get the first mode in case of multiple modes
        data[col].fillna(mode_val, inplace=True)

print("\nMissing Values after imputation:")
print(data.isnull().sum())
```

This code identifies missing values in categorical columns and replaces them with the mode (the most frequently occurring value).

```python
# Create new features:
# Price per square foot.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.preprocessing import LabelEncoder

warnings.filterwarnings('ignore')

data = pd.read_csv('/content/sample_submission.csv')

# Calculate Price per square foot
if 'GrLivArea' in data.columns and 'SalePrice' in data.columns:
    data['PricePerSqFt'] = data['SalePrice'] / data['GrLivArea']
else:
    print("Warning: 'GrLivArea' or 'SalePrice' column not found. Cannot calculate PricePerSqFt.")



```

This code creates a new feature called PricePerSqFt (Price per Square Foot) by dividing the Sale Price (SalePrice) by the Above-Ground Living Area (GrLivArea). This feature helps standardize pricing across different house sizes.

```python
#  Create new features:
# Age of the house (current year - year built).
# Assuming 'YearBuilt' column exists in the DataFrame
if 'YearBuilt' in data.columns:
    current_year = pd.Timestamp.now().year  # Get the current year
    data['HouseAge'] = current_year - data['YearBuilt']
else:
    print("Warning: 'YearBuilt' column not found. Cannot calculate HouseAge.")
```

This code calculates the age of each house by subtracting the year it was built (YearBuilt) from the current year. This feature helps in analyzing how the house’s age affects its value.

```python
#  Use feature importance from Random Forest to select key features.

from sklearn.ensemble import RandomForestRegressor

# Assuming 'data' DataFrame and 'SalePrice' target variable are defined
X = data.drop('SalePrice', axis=1)  # Features
y = data['SalePrice']  # Target variable

# Initialize and train a RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)  # Adjust parameters as needed
rf_model.fit(X, y)

# Get feature importances
feature_importances = rf_model.feature_importances_

# Create a DataFrame for better visualization
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print or visualize the feature importances
print(feature_importance_df)

# Select top N features (e.g., top 10)
top_n_features = 10  # Adjust as needed
selected_features = feature_importance_df['Feature'][:top_n_features].tolist()

# Create a new DataFrame with only the selected features and target variable
X_selected = data[selected_features]
X_selected
```

```python
#  Train the following models:
# Linear Regression
# Random Forest
# Gradient Boosting (XGBoost)

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize models
linear_model = LinearRegression()
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
xgb_model = XGBRegressor(n_estimators=100, random_state=42)

# Train models
linear_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
gb_model.fit(X_train, y_train)
xgb_model.fit(X_train, y_train)

# Make predictions
linear_predictions = linear_model.predict(X_test)
rf_predictions = rf_model.predict(X_test)
gb_predictions = gb_model.predict(X_test)
xgb_predictions = xgb_model.predict(X_test)

# Evaluate models
def evaluate_model(y_true, y_predicted, model_name):
    mse = mean_squared_error(y_true, y_predicted)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_predicted)
    print(f"{model_name}:")
    print(f"  Mean Squared Error (MSE): {mse}")
    print(f"  Root Mean Squared Error (RMSE): {rmse}")
    print(f"  R-squared (R2): {r2}")

evaluate_model(y_test, linear_predictions, "Linear Regression")
evaluate_model(y_test, rf_predictions, "Random Forest")
evaluate_model(y_test, gb_predictions, "Gradient Boosting")
evaluate_model(y_test, xgb_predictions, "XGBoost")
```

This process trains and evaluates four regression models to predict house prices based on selected features. Each model is assessed using standard performance metrics.

1. Models Used for Training
Four different models are trained on the training dataset (X_train, y_train):

Linear Regression – A simple model that assumes a linear relationship between features and target.
Random Forest Regressor – An ensemble learning method using multiple decision trees for better accuracy and reduced overfitting.
Gradient Boosting Regressor – A boosting technique that corrects prediction errors iteratively.
XGBoost Regressor – An optimized form of gradient boosting that improves speed and performance.

# 7. Model Optimization

1)Perform hyperparameter tuning using RandomizedSearchCV.


```python
# Perform hyperparameter tuning using RandomizedSearchCV.

from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid for Random Forest
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a RandomizedSearchCV object
rf_random_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=rf_param_grid,
    n_iter=10,  # Number of parameter settings that are sampled
    cv=5,  # Number of cross-validation folds
    scoring='neg_mean_squared_error',  # Use negative MSE as the scoring metric
    verbose=2,
    random_state=42,
    n_jobs=-1  # Use all available CPU cores
)


# Fit the RandomizedSearchCV object to the training data
rf_random_search.fit(X_train, y_train)

# Print the best hyperparameters and the best score
print("Best Hyperparameters:", rf_random_search.best_params_)
print("Best Score (negative MSE):", rf_random_search.best_score_)

# Evaluate the best model
best_rf_model = rf_random_search.best_estimator_
rf_predictions = best_rf_model.predict(X_test)
evaluate_model(y_test, rf_predictions, "Random Forest (Tuned)")
```

This step analyzes which features have the most impact on individual house price predictions using SHAP (SHapley Additive exPlanations) values. SHAP is a powerful technique for explaining machine learning model predictions by assigning importance to each feature.

1. What is SHAP?
SHAP values measure how much each feature contributes to a specific prediction.

Positive SHAP values → The feature increases the predicted price.
Negative SHAP values → The feature decreases the predicted price.
Higher absolute SHAP values → The feature has more influence on the prediction.
2. How SHAP is Used Here
A. Summarizing Feature Importance
The SHAP summary plot (bar chart) displays the overall influence of each feature.
Features with higher SHAP values have a greater impact on price predictions.
Helps identify key drivers of house prices (e.g., square footage, location, number of rooms).
B. Explaining Individual Predictions
The SHAP force plot shows how individual feature values affect a single house's predicted price.
It visually breaks down the push and pull of different features on a specific prediction.
Helps interpret why the model predicts a higher or lower price for a given house.
3. Why SHAP is Useful
Provides transparent, interpretable explanations for complex machine learning models.
Helps build trust in the model’s decisions.
Can guide feature selection by identifying which attributes are most impactful.
