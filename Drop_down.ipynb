```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go
import pandas as pd
```

```python
app = dash.Dash(__name__)
df = pd.DataFrame({
    'date': pd.date_range(start='1/1/2020', periods=10),
    'country': ['USA', 'Canada', 'Mexico', 'UK', 'Germany', 'France', 'Australia', 'Japan', 'China', 'India'],
    'confirmed': range(10, 110, 10),
    'deaths': range(1, 11),
    'recovered': range(5, 55, 5)
})

df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
app.layout = html.Div([
    dcc.Input(id='country-input', type='text', placeholder='Enter country'),
    dcc.Graph(id='covid-graph'),
    html.Div(id='error-message', style={'color': 'red'})
])

@app.callback(
    [Output('covid-graph', 'figure'), Output('error-message', 'children')],
    [Input('country-input', 'value')]
)
def update_graph(country):
    if not country:
        return go.Figure(), "Please enter a country."

    country = country.lower()
    country_data = df[df['country'].str.lower() == country]

    if country_data.empty:
        return go.Figure(), f"No data found for {country.title()}."

    aggregated_data = country_data.groupby('date').sum().reset_index()

    figure = go.Figure(data=[
        go.Bar(x=aggregated_data['date'], y=aggregated_data['confirmed'], name='Confirmed Cases', marker={'color': 'yellow'}),
        go.Bar(x=aggregated_data['date'], y=aggregated_data['deaths'], name='Deaths', marker={'color': 'red'})
    ])

    figure.update_layout(title=f"COVID-19 Data for {country.title()}", barmode='group')

    return figure, ""

if __name__ == '__main__':
    app.run_server(debug=True)
```
